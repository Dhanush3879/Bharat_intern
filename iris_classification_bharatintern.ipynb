```python
import matplotlib.pyplot as plt

fig, ax = plt.subplots()
ax.axis('equal')

l = ['Versicolor', 'Setosa', 'Virginica']
s = [50, 50, 50]

# Define custom colors
colors = ['pink', 'purple', 'violet']

# Create pie chart with custom colors
ax.pie(s, labels=l, autopct='%1.2f%%', colors=colors)

plt.show()
```

```python
import plotly.express as px

# Define custom colors for each species
custom_colors = {'Iris-versicolor': 'pink', 'Iris-setosa': 'purple', 'Iris-virginica': 'violet'}

fig = px.scatter(iris, x="sepal_width", y="sepal_length", color="species",
                 color_discrete_sequence=[custom_colors[species] for species in iris['species'].unique()])

fig.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Define custom colors for each species
custom_colors = {'Iris-versicolor': 'pink', 'Iris-setosa': 'purple', 'Iris-virginica': 'violet'}

# Set custom palette
custom_palette = [custom_colors[species] for species in iris['species'].unique()]

# Create pairplot with custom colors
sns.pairplot(iris, hue='species', palette=custom_palette)
plt.show()
```

```python
x = iris.drop("species", axis=1)
y = iris["species"]
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, 
                                                    test_size=0.2, 
                                                    random_state=0)

```

```python
from sklearn.neighbors import KNeighborsClassifier
knn = KNeighborsClassifier(n_neighbors=1)
knn.fit(x_train, y_train)
```

```python
#(sepal_length ,sepal_width ,petal_length ,petal_width)
x_new = np.array([[5, 2.9, 1, 0.2]])
prediction = knn.predict(x_new)
print("Prediction: {}".format(prediction))
```
